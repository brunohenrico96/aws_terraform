```python
from pyspark.sql import SparkSession
import os

# 1. Configuração do Spark
# Usando 6 núcleos conforme solicitado
spark = SparkSession.builder \
    .appName("Olist_Bronze_to_Silver") \
    .master("local[6]") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

# 2. Definição dos caminhos
# Como o notebook está na pasta 'data', os CSVs estão em 'bronze/'
pasta_bronze = "bronze/"
pasta_silver = "silver/"

arquivos = [f for f in os.listdir(pasta_bronze) if f.endswith('.csv')]

print(f"Encontrados {len(arquivos)} arquivos na pasta bronze.\n")

for arquivo in arquivos:
    # Limpando o nome da tabela para ficar mais amigável
    nome_tabela = arquivo.replace(".csv", "").replace("olist_", "").replace("_dataset", "")
    caminho_entrada = os.path.join(pasta_bronze, arquivo)
    caminho_saida = os.path.join(pasta_silver, nome_tabela)

    print(f"--- Processando: {nome_tabela} ---")
    
    # Leitura
    df = spark.read.csv(
        caminho_entrada,
        header=True,
        inferSchema=True
    )
    
    contagem = df.count()
    print(f"Linhas lidas: {contagem}")

    # Gravação em Parquet
    print(f"Convertendo para Parquet em: {caminho_saida}...")
    df.write \
        .mode("overwrite") \
        .parquet(caminho_saida)
    
    print(f"Sucesso! {nome_tabela} finalizado.\n")

print("--- Todas as conversões foram concluídas! ---")

spark.stop()
```
